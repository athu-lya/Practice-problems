```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mall_data=pd.read_csv('/content/Mall_Customers.csv')
```

```python
mall_data.head()
```

```python
mall_data.isna().sum()
```

```python
mall_data.describe()
```

```python
mall_data.info()
```

```python
mall_data
```

```python
from sklearn.cluster import KMeans
```

```python
X=mall_data.iloc[:,[3,4]].values
```

```python
X
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('number of clusters')
plt.ylabel('wcss')
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
```

```python
y_kmeans=kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 01')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label='Cluster 02')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='Cluster 03')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='yellow',label='Cluster 04')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],s=100,c='cyan',label='Cluster 05')
plt.title('Cluster of customers')
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score_avg=silhouette_score(X,y_kmeans)
silhouette_score_avg
```
